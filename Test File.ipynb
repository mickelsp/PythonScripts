We want to see what plotting looks like.

```python
a=[1,2,3]
```

```python
b=[3,1,6]
```

I'd like to use *Matlab* style syntax, so I can keep modifying the plot once I've sent the first command. First, I'll define some plotting parameters so that I can reuse them.

```python
fontsize=16
markersize=10
linewid=4
```

Now, I actually apply the *plot* command with some familiar-feeling syntax to produce my plot.

```python
line1=plot(a,b,'r--',linewidth=linewid)
line2=plot(b,a,'b-',linewidth=linewid)
xticklabels = getp(gca(), 'xticklabels')
yticklabels = getp(gca(), 'yticklabels')
setp(xticklabels, fontsize=fontsize, weight='bold')
setp(yticklabels, fontsize=fontsize, weight='bold')
ylabel('Some Numbers',fontsize=fontsize,fontweight='bold')
xlabel('The x Index',fontsize=fontsize,fontweight='bold')
legend(('First','Second'),loc='best')
grid()
show()
```
